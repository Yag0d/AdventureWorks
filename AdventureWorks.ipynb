#Como tem sido o crescimento do negócio ao longo do tempo? E nas suas principais linhas de produto?

```python
# Combinando tabelas de vendas
vendas = pd.concat([sales_2015, sales_2016, sales_2017])

# Convertendo a coluna OrderDate para datetime
vendas['OrderDate'] = pd.to_datetime(vendas['OrderDate'])

# Extraindo o ano de OrderDate
vendas['Year'] = vendas['OrderDate'].dt.year

# Unindo com os dados de produtos
vendas = vendas.merge(products, on='ProductKey', how='left')
vendas = vendas.merge(product_subcategories, on='ProductSubcategoryKey', how='left')
vendas = vendas.merge(product_categories, on='ProductCategoryKey', how='left')

# Agrupa do vendas por ano e por categoria de produto
vendas_por_ano = vendas.groupby('Year', as_index=False)['OrderQuantity'].sum()
vendas_por_produto = vendas.groupby(['CategoryName', 'Year'], as_index=False)['OrderQuantity'].sum()

# Gráfico do crescimento ao longo do tenpo
fig = px.line(vendas_por_ano, x='Year', y='OrderQuantity', title='Crescimento do Negócio ao Longo do Tempo')
fig.show()

# Gráfico do crescimento por linha de produto
fig = px.line(vendas_por_produto, x='Year', y='OrderQuantity', color='CategoryName', title='Crescimento das Linhas de Produto ao Longo do Tempo')
fig.show()
```

#Qual é a região que mais vende em volume total de vendas? E em volume de itens?


```python
# Unindo os dados de vendas com os dados de territórios
vendas = vendas.merge(territories, left_on='TerritoryKey', right_on='SalesTerritoryKey', how='left')

# Agrupandi vendas por região
vendas_por_regiao = vendas.groupby('Region', as_index=False).agg({'OrderQuantity': 'sum', 'OrderLineItem': 'count'})

# Gráfico de volume total de vendas por região
fig = px.bar(vendas_por_regiao, x='Region', y='OrderLineItem', title='Volume de Itens Vendidos por Região')
fig.show()

# Gráfico de volume de itens por região
fig = px.bar(vendas_por_regiao, x='Region', y='OrderQuantity', title='Volume Total de Vendas por Região ')
fig.show()
```

#Quais são os produtos campeões de vendas e mais lucrativos?

```python
# Agrupando vendas por produto
vendas_por_produto = vendas.groupby(['ProductName'], as_index=False).agg({'OrderQuantity': 'sum', 'ProductPrice': 'sum'})

# Produtos campeões de vendas
produtos_mais_vendidos = vendas_por_produto.nlargest(5, 'OrderQuantity')
fig = px.bar(produtos_mais_vendidos, x='ProductName', y='OrderQuantity', title='Top 5 Produtos Campeões de Vendas')
fig.show()

# Produtos mais lucrativos
produtos_mais_lucrativos = vendas_por_produto.nlargest(5, 'ProductPrice')
fig = px.bar(produtos_mais_lucrativos, x='ProductName', y='ProductPrice', title='Top 5 Produtos Mais Lucrativos')
fig.show()
```

#Qual região deve ser priorizada para abertura do novo escritório? (Australia em primeiro lugar no gráfico. Porém observamos a crescente do Canadá, que também poderá vir a ser uma região para priorização.

```python
# Combinando tabelas de vendas
vendas = pd.concat([sales_2015, sales_2016, sales_2017])

# Convertendo a coluna OrderDate para datetime
vendas['OrderDate'] = pd.to_datetime(vendas['OrderDate'])

# Extraindo o ano de OrderDate
vendas['Year'] = vendas['OrderDate'].dt.year

# Removendo colunas desnecessárias de territories antes de fazer o merge
territories = territories[['SalesTerritoryKey', 'Region']]

# Unindo os dados de vendas com os dados de territórios
vendas = vendas.merge(territories, left_on='TerritoryKey', right_on='SalesTerritoryKey', how='left')

# Agrupando vendas por região e ano
vendas_por_regiao_ano = vendas.groupby(['Region', 'Year'], as_index=False).agg({'OrderQuantity': 'sum', 'OrderLineItem': 'count'})

# Gráfico de crescimento das vendas por região ao longo dos anos
fig = px.line(vendas_por_regiao_ano, x='Year', y='OrderQuantity', color='Region', title='Crescimento das Vendas por Região ao Longo dos Anos')
fig.show()
```

#Análise de meta

```python
vendas['Vendas_total'] = vendas['OrderQuantity'] * vendas['ProductPrice']
```

```python

# Agrupando vendas por ano
vendas_por_ano = vendas.groupby('Year', as_index=False)['Vendas_total'].sum()

# Calculando a média dos três primeiros anos
media_primeiros_3_anos = vendas_por_ano[vendas_por_ano['Year'] <= 2017]['Vendas_total'].mean()

# Defininindo metas a partir do 4º ano com crescimento de 20% a cada biênio
metas = []
anos = sorted(vendas_por_ano['Year'].unique())
for i, ano in enumerate(anos):
    if i < 3:
        metas.append(media_primeiros_3_anos)
    else:
        metas.append(metas[-2] * 1.2)

# Adicionando as metas ao DataFrame de vendas por ano
vendas_por_ano['Meta'] = metas

# Verificando quais anos atingiram a meta
vendas_por_ano['AtingiuMeta'] = vendas_por_ano['Vendas_total'] >= vendas_por_ano['Meta']

# Gráfico comparativo de vendas e metas
fig = px.line(vendas_por_ano, x='Year', y=['Vendas_total', 'Meta'],
              labels={'value': 'Valor de Vendas', 'variable': 'Legenda'},
              title='Comparação das Vendas com as Metas Estabelecidas')
fig.show()
```

```python
# Imprimindo os anos em que a meta foi atingida e o valor de vendas
anos_atingiram_meta = vendas_por_ano[vendas_por_ano['AtingiuMeta'] == True]
print("Anos em que a meta foi atingida:")
print(anos_atingiram_meta[['Year', 'Vendas_total', 'Meta']])
```
